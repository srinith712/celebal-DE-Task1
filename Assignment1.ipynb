```python
df["final_price"] = (
    df["final_price"]
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)
```

```python
df["final_price"] = pd.to_numeric(
    df["final_price"],
    errors="coerce"
)
```

```python
df["final_price"] = df["final_price"].fillna(
    df["final_price"].mean()
)
```

```python
df["initial_price"] = (
    df["initial_price"]
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)
```

```python
df["initial_price"] = pd.to_numeric(
    df["initial_price"],
    errors="coerce"
)
```

```python
df["initial_price"] = df["initial_price"].fillna(
    df["initial_price"].mean()
)
```

Creating a new column based on available columns
